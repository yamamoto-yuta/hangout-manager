# バイナリパターンの学習データセットを作成

```python
df_train_bin = df_label.iloc[:, :3]
```

```python
df_train_bin.head()
```

```python
def map_2vec(row):
    vector = np.zeros(len(df_genre) - 1).astype(np.int64)
    for x in row:
        if x != 8:
            vector[x] = 1
    return vector / sum(vector)

df_train_bin.apply(map_2vec, axis=1)
```

```python
# CSV出力
df_csv_bin = df_input.copy()
df_csv_bin['label'] = df_train_bin.apply(map_2vec, axis=1)
df_csv_bin.to_csv('/home/res/datasets/static-info_data_bin.csv', index=False)

df_csv_bin
```

# 重みパターンの学習データセットを作成

```python
# セル内のリストが文字列になっているのでリストに変換
def map_stol(x):
    lst = x.replace('[', '').replace(']', '').split(',')
    return [int(v) for v in lst]

df_label.iloc[:, 3] = df_label.iloc[:, 3].map(map_stol)
```

```python
def apply_rm_duplicate(row):
    for i in range(3):
        # 途中でリストが空になったらNoneを返す
        if not row[3]:
            return None
        # 第1~3希望で選んだジャンルがリストに含まれていたら削除
        if row[i] in row[3]:
            row[3] = row[3].remove(row[i])
    
    # リストが空or"なし"のみの場合Noneを返す
    if not row[3] or (len(row[3]) == 1 and row[3][0] == 8):
        return None
    
    return row[3]
            
df_label.iloc[:, 3] = df_label.apply(apply_rm_duplicate, axis=1)
```

```python
# なしをNoneに変換
df_label.iloc[:, 1] = df_label.iloc[:, 1].map(lambda x: None if x == 8 else x)
df_label.iloc[:, 2] = df_label.iloc[:, 2].map(lambda x: None if x == 8 else x)
```

```python
df_label.head()
```

```python
def map_calc_weight(row):
    weight = np.array([1.0, 0.7, 0.4, 0.1])
    vector = np.zeros(len(df_genre) - 1).astype(np.float64)
    
    # 第1~3希望まで重みを計算
    for i, x in enumerate(row[:3]):
        if not np.isnan(x):
            vector[int(x)] = weight[i]
    
    # 第4希望の重みを計算
    if row[3] is not None:
        for x in row[3]:
            vector[int(x)] = weight[-1]
    
    return vector / sum(vector)

df_label.apply(map_calc_weight, axis=1)
```

```python
# CSV出力
df_csv_weight = df_input.copy()
df_csv_weight['label'] = df_label.apply(map_calc_weight, axis=1)
df_csv_weight.to_csv('/home/res/datasets/static-info_data_weight.csv', index=False)

df_csv_weight
```
